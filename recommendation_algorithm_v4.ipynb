##### Refer to 'model_progression.ipynb' to find best model metadata to load

```python
# Initial params
split = 0.8
recurrent = 120


# Load model
model_source = '327-3_cl-0.5_idc-1_mid-0.5_ep-15_bz-2048_ldtv-2_sbuffs-1_gen-1_1-7'
model_source = '327-9_cl-0.005_idc-1_mid-0.9_ep-40_bz-2048_ldtv-3_sbuffs-1_gen-1_28-0' # 0.6459 / 0.6173
model_source = '327-9_cl-0.1_idc-1_mid-0.9_ep-40_bz-2048_ldtv-3_sbuffs-1_gen-1_7-5' # 0.6371 / 0.6281

ROOT_PATH = './'

model_source_path = core.os.path.join(ROOT_PATH, f'models/preloaded/{model_source}')
model = core.tf.keras.models.load_model(f'{model_source_path}/model')
```

```python
# Define selected tickers to be analyzed
ticker_groups = ['LSIP','BWPT','SKBM','BUKK','ACES','VIVA','TCPI','INDF','ICBP','TLKM','CMNP']
# Ticker group 327_9
ticker_groups = ['WEGE','ACES','MYOR','AMRT','AKRA','MEDC','AALI','PRDA','GJTL']
ticker_groups_db = [i + '.JK' for i in ticker_groups]
ticker_groups_expanded = core.np.expand_dims(core.np.array(ticker_groups), axis=1)
# ticker_groups_db_expanded = core.np.expand_dims(core.np.array(ticker_groups_db), axis=1)

# Fetch data version that used to train the model
benchmark_db = 'E:\\#PROJECT\\idx\\db\\v8\\dataset\\vid\\AALI.JK_shift0_interval5_recurrent120_split0.8.hdf5'
with core.h5py.File(benchmark_db, 'r') as f:
    data_version = f.attrs['data_version']

# Fetch standard scaler 
standard_scaler = []
for i, ticker_group_expanded in enumerate(ticker_groups_expanded):
    train_mean, train_std = core.standard_scaler_one_shot(ticker_group_expanded, data_version, ROOT_PATH=ROOT_PATH, split=split)
    standard_scaler.append([train_mean, train_std])
```

```python
# before: 0,1,2,3 if the db updated after the market closed
shifts_from_last = [0,1,2,3]

# Make predictions
db_path = './indicators_db/id_indicators.db'
db_conn = core.sqlite3.connect(db_path)

prediction_results = []
for shift_from_last in shifts_from_last:
    prediction_result = []
    for j, sticker in enumerate(ticker_groups_db):
        tomorrow_input = core.make_one_shot_input(sticker, db_conn, recurrent, shift_from_last=shift_from_last)
        tomorrow_input_n = (tomorrow_input - standard_scaler[j][0]) / standard_scaler[j][1]
        tomorrow_input_e = core.np.expand_dims(tomorrow_input_n, 0)
        tomorrow_prediction = model.predict(tomorrow_input_e)
        # val_acc = histories[i].history['val_accuracy'][-1]
        # print(sticker, f'val_acc:{val_acc*100:.0f}%',tomorrow_prediction)

        # Convert prediction to text
        tomorrow_prediction_argmax = core.np.argmax(tomorrow_prediction)
        prediction = 'UP' if tomorrow_prediction_argmax == 1 else 'DOWN'
        # print(sticker, prediction)
        prediction_result.append(prediction)
    prediction_results.append(prediction_result)

# Check double down and double up
up = []
down = []
for i, results in enumerate(prediction_results[0]):
    #  == prediction_results[2][i]
    if prediction_results[0][i] == prediction_results[1][i]:
        if results == 'UP':
            up.append(ticker_groups[i])
        elif results == 'DOWN':
            down.append(ticker_groups[i])
print(f'up: {up}')
print(f'down: {down}')
```

```python
import coremlv2 as core
import importlib
importlib.reload(core)
```

```python
# Initialize
base_path = './db/v7/dataset/v2/'
```

```python
# Shuffle tickers
tickers = core.us_db_stocklist(db='us_indicators.db')
excluded_stock = core.us_excluded_stock()
tickers = [ticker for ticker in tickers if ticker not in excluded_stock]
core.random.seed(0)
core.random.shuffle(tickers)
```

```python
# Rename tickers
base_filename = '_shift0_interval1_recurrent120_split0.8.hdf5'
target_identifier = 'MERGED-unshuffled'
for i, ticker in enumerate(tickers):
    source_filename = core.os.path.join(base_path, f'{ticker}{base_filename}')
    if core.os.path.exists(source_filename):
        target_filename = core.os.path.join(base_path, f'{i}{target_identifier}{base_filename}')
        core.os.rename(source_filename, target_filename)
```
